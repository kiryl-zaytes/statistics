Find which sentences in provided text "cats.txt" have more in common with first one. All sentences are about 'cats' but in some cases those are pets, in some - linux command 'cat' and also it could be os version of MaC.

```python
import scipy
import numpy as np
import re
import collections
import scipy.spatial.distance as dist
```

```python
lines = []
with open("cats.txt", "r") as f:
    lines = f.readlines()
print(lines[0])
```

```python
lines = [re.split('[^a-z]', l.lower()) for l in lines]
print(lines[0])
```

```python
list_of_words = [collections.Counter(l) for l in lines]
for i in list_of_words:
    i.pop('', None)
```

```python
list_of_words
```

```python
unique_words = set()
for i in list_of_words:
    unique_words |= i.keys()
```

```python
words_matrix = []
for sentence in list_of_words:
    for u_w in unique_words:
        words_matrix.append(sentence.get(u_w, 0))
```

```python
words_matrix = np.array(words_matrix).reshape((len(lines), len(unique_words)))
```

```python
koeff = []
for r in words_matrix:
    koeff.append((dist.cosine(words_matrix[0], r)))
```
